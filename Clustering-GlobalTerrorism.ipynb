```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd

# Assuming you already have your DataFrame 'df' and 'scaled_data' from the previous steps

# Apply PCA
pca = PCA()
data_pca = pca.fit_transform(scaled_data)

# Define range of clusters
cluster_range = range(1, 10)
cluster_errors = []

# Loop through each number of clusters
for num_clusters in cluster_range:
    clusters = KMeans(num_clusters, n_init=10, max_iter=100)
    clusters.fit(data_pca)
    
    # Append inertia (within-cluster sum of squares) to cluster_errors list
    cluster_errors.append(clusters.inertia_)
    
# Create DataFrame to visualize cluster errors
cluster_errors_df = pd.DataFrame({'num_clusters': cluster_range, 'Error': cluster_errors})
print(cluster_errors_df)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

best_score = -1
best_k = None

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, n_init=15, random_state=2345)
    kmeans.fit(data_pca)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(data_pca, labels)
    
    if silhouette_avg > best_score:
        best_score = silhouette_avg
        best_k = k

print("Best Silhouette Score:", best_score)
print("Optimal number of clusters (k):", best_k)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler



# Select relevant numerical features for clustering
# You may need to adjust the feature selection based on the dataset
features = ['iyear', 'imonth', 'iday', 'country', 'region', 'latitude', 'longitude', 'nkill', 'nwound']
df_selected = df[features].dropna()

# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(df_selected)
```

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture

# Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Apply Gaussian Mixture Model (GMM) clustering
gmm = GaussianMixture(n_components=5, random_state=42)
gmm_labels = gmm.fit_predict(X)
```

```python
from sklearn.metrics import classification_report, adjusted_rand_score

# Use 'country' as true labels for demonstration purposes
true_labels = df_selected['country'].values

# Function to evaluate and print classification report
def evaluate_clustering(true_labels, pred_labels, model_name):
    print(f"Classification Report for {model_name}:\n")
    print(classification_report(true_labels, pred_labels))
    print(f"Adjusted Rand Index: {adjusted_rand_score(true_labels, pred_labels)}\n")

# Evaluate KMeans
evaluate_clustering(true_labels, kmeans_labels, "KMeans")

# Evaluate DBSCAN
evaluate_clustering(true_labels, dbscan_labels, "DBSCAN")

# Evaluate GMM
evaluate_clustering(true_labels, gmm_labels, "GMM")
```
